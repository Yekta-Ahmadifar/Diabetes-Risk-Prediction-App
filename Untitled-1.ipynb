**Importing required libraries**

```python
import pandas as pd
import numpy as np
#import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score 
from sklearn.metrics import confusion_matrix , classification_report,  ConfusionMatrixDisplay 
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.pipeline import Pipeline 

```

**Exploratory data analysys**

```python
#load dataset
df=pd.read_csv("diabetes-pima.csv")

#exploratory data analysis
df.info()
df.duplicated().sum()
```

```python
df.head()
```

**Data Processing**

```python
#Replacing 0s with NaN
fix_cols = ['Glucose' , 'BloodPressure' , 'SkinThickness' , 'Insulin' , 'BMI']
df[fix_cols] = df[fix_cols].replace(0, np.nan)

#imput missing Values with Median
for col in fix_cols:
    df[col] = df[col].fillna(df[col].median())

#separate Features and Traget
X = df.drop(columns='Outcome', axis=1)
y = df['Outcome']   

#Feature scaling 
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X) 

#Test-Train split
x_train , x_test , y_train , y_test = train_test_split(X_scaled , y, test_size=0.2 , random_state=42)
```

**Logistic Regression Model**

```python
#initialize and train model
LogR = Pipeline([
('scaler' , StandardScaler()),
('model' , LogisticRegression(max_iter=500 , class_weight='balanced' , random_state=42))
])

LogR.fit(x_train , y_train)

#making predictions
y_pred_LogR = LogR.predict(x_test)

#Evaluation metrics
LRAcc= accuracy_score(y_test, y_pred_LogR) 
LRRecall = recall_score(y_test , y_pred_LogR)
LRPerc = precision_score(y_test , y_pred_LogR)

#evaluation for the model
print(f"Accuracy Score: {accuracy_score(y_test, y_pred_LogR):.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred_LogR))
```

Feature importance

```python
coef = LogR.named_steps['model'].coef_[0]
importanceLR = pd.Series(coef , index=X.columns).sort_values()
importanceLR.plot(kind='barh' , title="Logistic Regression feature importance")
```

**Random Forest model**

```python
#initializing model with 100 individual trees
RF = RandomForestClassifier(n_estimators=100 , class_weight='balanced' ,  random_state=42)

#training, no scaling required for RF model
RF.fit(x_train , y_train)

#make predictions 
y_pred_RF = RF.predict(x_test)

#Evaluation metrics
RFAcc= accuracy_score(y_test, y_pred_RF) 
RFRecall = recall_score(y_test , y_pred_RF)
RFPerc = precision_score(y_test , y_pred_RF)

#evaluation for the model
print(f"Accuracy Score: {RFAcc:.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred_RF))
```

Feature importance


```python
RFimportance = RF.feature_importances_
feature_nameRF = X.columns

#creat plot
RFimportance = pd.Series(RFimportance, index=feature_nameRF)
RFimportance.nlargest(10).plot(kind="barh")
plt.title("Random Forest Feature Importance")
plt.show()
```

**Decision Tree Model**

```python
#initializing the model
DT = DecisionTreeClassifier(max_depth=5 , class_weight='balanced' ,  random_state=42)

#train
DT.fit(x_train , y_train)

#prediction
y_pred_DT = DT.predict(x_test)

#Evaluation metrics
DTAcc= accuracy_score(y_test, y_pred_DT) 
DTRecall = recall_score(y_test , y_pred_DT)
DTPerc = precision_score(y_test , y_pred_DT)

#evaluation 
print(f"Decision Tree Accuracy:{DTAcc:.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred_DT))
```

Feature importance

```python
DTimportance = DT.feature_importances_
feature_nameDT = X.columns

#plotting
DTimportance = pd.Series(DTimportance , index=feature_nameDT)
DTimportance.nlargest(5).plot(kind="barh")
plt.title("Decision Tree Feature importance")
plt.show() 
```

**Naive Bayes**

```python
#initializing Model
NB = GaussianNB ()
NB.fit(x_train , y_train)

#prediction
y_pred_NB = NB.predict(x_test)

#Evaluation metrics
NBAcc = accuracy_score(y_test ,y_pred_NB)
NBRecall = recall_score(y_test , y_pred_NB)
NBPerc = precision_score(y_test , y_pred_NB)

#evaluation
print(f" Naive Bayes:{NBAcc:.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred_NB))
```

**KNN model** 

```python
#scaling data
scaler = StandardScaler()
x_train_scale = scaler.fit_transform(x_train)
x_test_scale = scaler.transform(x_test)

#initilize and train
KNN = KNeighborsClassifier(n_neighbors=12)
KNN.fit(x_train_scale , y_train)

#prediction
y_pred_KNN = KNN.predict(x_test_scale)

#Evaluation metrics
KNNAcc = accuracy_score(y_test ,y_pred_KNN)
KNNRecall = recall_score(y_test , y_pred_KNN)
KNNPerc = precision_score(y_test , y_pred_KNN)


#evaluation
print(f"KNN Accuracy: {KNNAcc:.4f}")
print(classification_report(y_test , y_pred_KNN)) 
```

**Super Vector Machine**

```python
SVM = Pipeline([
    ('scaler' ,StandardScaler()) ,
    ('moddel' , SVC(
        kernel= 'rbf' , 
        C=1.0 ,
        gamma='scale' ,
        class_weight = 'balanced' ,
        random_state=42
    ))
])

SVM.fit(x_train , y_train)
y_pred_svm = SVM.predict(x_test)
svmAcc = accuracy_score(y_test , y_pred_svm)
svmRecall = recall_score(y_test , y_pred_svm)
svmPerc = precision_score(y_test , y_pred_svm)

print(f"KNN Accuracy: {svmAcc:.4f}")
print(classification_report(y_test , y_pred_svm)) 
```

**Gradient Boosting**

```python
GB = GradientBoostingClassifier(
    n_estimators=200,
    learning_rate=0.05,
    max_depth=3,
    random_state=42
)

GB.fit(x_train, y_train)

#prediction
y_pred_GB = GB.predict(x_test)

#Evaluation 
GBAcc = accuracy_score(y_test , y_pred_GB)
GBRecall = recall_score(y_test , y_pred_GB)
GBPerc = precision_score(y_test , y_pred_GB)

print(f"Gradient Boosting Accuracy: {GBAcc:.4f}")
print("\nclassification Report:")
print(classification_report(y_test , y_pred_GB))
```

**Model comparison**

Comparison Graph

```python
models = ["LR" , "RF" , "DT" , "NB" , "KNN" , "SVM" , "GB"]
Accuracy = [LRAcc , RFAcc , DTAcc , NBAcc , KNNAcc , svmAcc , GBAcc]
Percision = [LRPerc , RFPerc , DTPerc , NBPerc , KNNPerc , svmPerc, GBPerc ]
Recall = [LRRecall , RFRecall , DTRecall , NBRecall , KNNRecall , svmRecall , GBRecall]

XG = np.arange(len(models))
width = 0.25

plt.figure(figsize=(12 , 7))
plt.bar(XG - width, Accuracy , width, label='Accuracy' , color="blue" )
plt.bar(XG , Recall , width, label='Recall' , color="Red" )
plt.bar(XG + width ,Percision , width, label='Percision' , color="green" )


plt.title('Model comparison' , fontsize=16 , fontweight='bold')
plt.ylabel("Score" , fontsize=14)
plt.xticks(XG , models)
plt.ylim(0, 1.0)
plt.legend(loc='best' , frameon=True)
plt.grid(axis='y', linestyle='--' , alpha=0.7 )
plt.tight_layout()
plt.show
```
